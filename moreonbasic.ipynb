```python
l=[1,2,3,4,4,"sudh","kumar",324,34.456,"abc"]
```

```python
num=[]
strr=[]
flt=[]
for i in l :
  if type(i)== int:
    num.append(i)
  elif type(i) == float:
    flt.append(i)
  else:
    strr.append(i)
print(num)
print(strr)
print(flt)
```
